LEFT JOIN gives all rows from the left table and only those rows from right table where the values are common in the columns used in left join condition

```python
ps.sqldf("""

SELECT
      C.City,
      df_dataset.Embarked,
      df_dataset.Sex,
    
      AVG(df_dataset.Age) AS avg_age,
      SUM(df_dataset.Age) AS tot_age,
      MIN(df_dataset.Age) AS min_age,
      MAX(df_dataset.Age) AS max_age,
      COUNT(df_dataset.PassengerId) AS tot_PassengerId,
      COUNT(DISTINCT df_dataset.Ticket) AS Dist_Ticket

FROM df AS df_dataset

LEFT JOIN cities AS  C
      ON df_dataset.Embarked=C.code

GROUP BY df_dataset.Embarked,
    df_dataset.Sex



ORDER  BY df_dataset.Embarked DESC,
        df_dataset.Sex ASC

LIMIT 5

""")
```

# ***Ranking in SQL***
There are lot times when you would want to sort your data and give a rank to each row in the dataset. RANK(), DENSE_RANK() and ROW_NUMBERS() are three functions which help you do just that. Generally these functions are used with OVER() function to sort the data based on some column before giving rank to each row and also to partition the data, which we will learn later. Let's explore RANK() first.




 RANK() provides a rank to each row. If any n rows have the same value in the sorting column, then RANK() gives the same value to all of these rows. For the next row where the value is different than the previous n rows, the rank given will be previous rank plus n. Here is an example -

difference-between-rank-and-dense_rank-functio.png

download (4).png

images.png

```python
#RANK() sorts the data, gives the same rank to all consecutive n rows with same values, 
#but for next row with a different value, it gives a rank equal to previous rank plus n
ps.sqldf("""
SELECT 
      df_dataset.Name,
      df_dataset.Embarked,
      df_dataset.Sex,
      df_dataset.Age,
      df_dataset.Fare,
      RANK() OVER(ORDER BY df_dataset.Age DESC) AS rnk
FROM df AS df_dataset
""")
```

As you can see, "OVER(ORDER BY df.Age DESC)" orders Age in descending order and then Rank() is used to assign a rank to each row. Notice row 3 and 4, where the age values are the same(i.e. 71) and hence both rows have the same rank(=3). In row 5, the ranking begins from 5(=previous rank 3 plus 2) and not 4.




So by using RANK() there are some ranks which get skipped if there are same values in the sorting order. Using DENSE_RANK() we can ensure that no ranks are skipped. Notice in the example below where row 5 has a rank 4 and not 5.

```python
#DENSE_RANK() does not skip ranks like RANK() does
ps.sqldf("""

SELECT 
    df_dataset.Name,
    df_dataset.Embarked,
    df_dataset.Sex,
    df_dataset.Age,
    df_dataset.Fare,
    DENSE_RANK() OVER(ORDER BY df_dataset.Age DESC) AS rnk
    
FROM df AS df_dataset

""")
```

And then we have ROW_NUMBER() which gives different rank to each row even if the sorted values are same. Notice below how in row 3 and 4 age is same but the ranks are different.



```python
#ROW_NUMBER() sorts the data and gives different ranks to each row 
ps.sqldf("""

SELECT 
    data_df.Name,
    data_df.Embarked,
    data_df.Sex,
    data_df.Age,
    data_df.Fare,
    ROW_NUMBER() OVER(ORDER BY data_df.Age DESC) AS rnk
    
FROM df AS data_df

""")
```

```python

#You can use multiple columns to sort the data within OVER() function
ps.sqldf("""

SELECT 
    data_df.Name,
    data_df.Embarked,
    data_df.Sex,
    data_df.Age,
    data_df.Fare,
    RANK() OVER(ORDER BY data_df.Age DESC, data_df.Fare ASC) AS rnk
    
FROM df AS data_df

""")
```

Further, OVER() also supports partitioning of data. Suppose you want to give separate ranks withtin specific levels of data, then you can use "PARTITION BY" in OVER() function. In the example below, for each combination of df.Embarked and df.Sex, we will get different sets of ranks.


***over()clause=***defines how to partition and order rows of table, which is to be processed by window function rank().

***dense_rank()*** is a window function, which will assign rank in ordered partition of data rank columns. If two hackers have same scores then they will be assigned same rank.


https://www.geeksforgeeks.org/mysql-partition-by-clause/

partition by clause will partition table into groups that are having same challenge_id.

```python
#Using PARTITION BY to create buckets for giving separate sets of ranks
ps.sqldf("""

SELECT 
    data_df.Name,
    data_df.Embarked,
    data_df.Sex,
    data_df.Age,
    data_df.Fare,
    RANK() OVER(PARTITION BY data_df.Embarked, data_df.Sex 
                ORDER BY data_df.Age DESC, data_df.Fare DESC) AS rnk
    
FROM df AS data_df

""")
```

# ***Lead and Lag***
LEAD() and LAG() can be used to get the nth previous or nth next value in a column, after sorting the data using OVER()

https://www.geeksforgeeks.org/sql-server-lead-function-overview/

https://www.geeksforgeeks.org/sql-server-lag-function-overview/

Note:The LEAD() and LAG() function are always used with OVER(). Missing of an over clause will raise an error.

```python
#Examples for LEAD() and LAG()
ps.sqldf("""

SELECT 
    data_df.Name,
    LEAD(data_df.Name, 1) OVER(ORDER BY data_df.Name) AS next_name,
    LEAD(data_df.Name, 2) OVER(ORDER BY data_df.Name) AS second_next_name,
    LAG(data_df.Name, 1) OVER(ORDER BY data_df.Name) AS previous_name,
    LAG(data_df.Name, 2) OVER(ORDER BY data_df.Name) AS second_previous_name
    
 
    
FROM df AS data_df

""")
```

# ***Subqueries***
Subqueries are small queries which are used within a bigger main SQL query. Subqueries store the output of a SQL temporarily and more SQL queries can be written on top of these.
Here are some examples of Subquery -

```python

#Using subqueries to get average age and then filtering only those rows where average age is below 30
ps.sqldf("""

SELECT 

    df1.Embarked,
    df1.avg_age
    
FROM (SELECT 
            a.Embarked, 
            AVG(a.Age) AS avg_age 
        FROM df AS a
        GROUP BY a.Embarked) AS df1

WHERE df1.avg_age<30
    
""")
```

Here df1 is a temporary datasest which gets created via a subquery

```python


#Using subqueries to get average age and number of passengers separately and then joining the outputs
ps.sqldf("""

SELECT 
    df1.Embarked,
    df1.avg_age,
    df2.tot_passengers
FROM (SELECT 
            a.Embarked, 
            AVG(a.Age) AS avg_age 
        FROM df AS a
        GROUP BY a.Embarked) AS df1

INNER JOIN (SELECT 
                b.Embarked, 
                COUNT(b.PassengerID) AS tot_passengers
            FROM df AS b
            GROUP BY b.Embarked) AS df2
    ON df1.Embarked = df2.Embarked
    
""")
```

You can nested subqueries inside other subquery and also use them to filter dataset as seen below -

```python


#Using nested subqueries to filter data
ps.sqldf("""
SELECT b.* FROM df AS b
WHERE b.Embarked IN 
       (SELECT 
            DISTINCT df1.Embarked
        FROM (SELECT 
                    a.Embarked, 
                    AVG(a.Age) AS avg_age 
                FROM df AS a
                GROUP BY a.Embarked) AS df1
        WHERE df1.avg_age>30)
    
""")
```
